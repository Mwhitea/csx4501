```python
# Assign the dataframe values from the previous cell into a new dataframe variable
# and make a bar plot with the text values horizontally and the numeric values as the bar heights
dfnew = pd.DataFrame(df)

ax= dfnew.iloc[0:10].plot(kind='bar', 
                       x='Year', 
                       y='Production.Coal', 
                       figsize=(14,7),
                       color=(0.5,0,0.87,0.99))
ax.legend(bbox_to_anchor=(1, 1))
ax.set_xlabel('Year',fontsize=10)
ax.set_ylabel('Production of Coal',fontsize=10)
```

```python
# Re-do the plot from the previous cell as a horizontal bar plot
dfnew = pd.DataFrame(df)

ax= dfnew.iloc[0:10].plot(kind='barh', 
                       x='Year', 
                       y='Production.Coal', 
                       figsize=(14,7),
                       color=(0.5,0,0.87,0.99))
ax.legend(bbox_to_anchor=(1, 1))
ax.set_xlabel('Production of Coal',fontsize=10)
ax.set_ylabel('Year',fontsize=10)
```

```python
# Re-do the plot from the previous cell
# and change at least two aesthetic elements (colors, labels, titles, ...)

dfnew = pd.DataFrame(df)

ax= dfnew.loc[0:10].plot(kind='barh', 
                       x='Year', 
                       y='Production.Coal', 
                       figsize=(14,5),
                       color=(0.4,1,0.17,0.99))
ax.legend(bbox_to_anchor=(1, 1))
ax.set_xlabel('Production of Coal',fontsize=20)
ax.set_ylabel('Specific Year',fontsize=20)
```
